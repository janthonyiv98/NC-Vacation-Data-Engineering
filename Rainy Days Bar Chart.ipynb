```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib notebook
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
from scipy.stats import linregress
from scipy import stats
from datetime import date
```

```python
combined_data = pd.read_csv("combined_data.csv")
nc_coordinate_data = pd.read_csv("NC_Coordinate_Data.csv")
```

```python
combined_df = pd.merge(combined_data, nc_coordinate_data, on="city_id")
combined_df.head()
```

```python
del combined_df['City']
combined_df = combined_df.rename(columns={'name':'city'})
combined_df['datetime'] = pd.to_datetime(combined_df['datetime'])
```

```python
rain_df = combined_df[combined_df['icon'].str.contains('rain')]
```

```python
rain_days_df = pd.DataFrame(rain_df['city'].value_counts())
rain_days_df
```

```python
x_axis = np.arange(len(rain_days_df))
plt.bar(x_axis, rain_days_df['count'])
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Cherokee", "Ocracoke", "Raleigh", "Nags Head", "Banner Elk", "Emerald Isle", "High Point", "Asheville", "Charlotte", "Boone", "Wilmington", "Winston-Salem"], rotation="vertical")
plt.title('Number of Rainy Days in North Carolina Cities (2023)')
plt.xlabel('City')
plt.ylabel('Rainy Days')
plt.show()
```
